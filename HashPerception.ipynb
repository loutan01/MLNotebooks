```python
def make_hash(img):
    
    rows,cols=img.shape
    rhash = []
    chash = []
    
    
    for row in range(rows):
        for col in range(cols-1):
            if img[row,col+1]>=img[row,col]:
                rhash.append(1)
            else:
                rhash.append(0)
    
    for row in range(rows):
        for col in range(cols-1):
            if img[row+1,col]>=img[row,col]:
                rhash.append(1)
            else:
                rhash.append(0)
    
    
    return int("".join([str(x) for x in rhash] + [str(x) for x in chash]),2)



make_hash(thumb)
```

```python
#take image, put to grayscale, HOG it
#hog: divide imige into 16x16 blocks
```

```python

#http://scikit-image.org/docs/dev/api/skimage.feature.html#skimage.feature.hog
#http://scikit-image.org/docs/dev/auto_examples/features_detection/plot_hog.html
def hogIt(image):
    fd, hog_image = hog(image, orientations=9, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualise=True)
    
    return exposure.rescale_intensity(hog_image, in_range=(0, 0.02))
```
